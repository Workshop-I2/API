```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
# Charger le modèle HateBERT et son tokenizer depuis Hugging Face
model_name = "GroNLP/hateBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
# Définir le texte à analyser (exemple de post de réseau social)
texts = [
    "I hate everyone in this place!",
    "You are such a nice person!",
    "All these people are disgusting and should leave.",
    "Gays should not be allowed",
    "Gays are nice people",
    "All black people must die",
    "All white people must die",
    "I love mexican food",
    "I love mexicans",
    "Fuck you bitch",
    "You talking to me mother fucker",
    "Listen to me you litle monkey",
    "I want to lick your pussy and then fuck you",
    "you blakc bastard",
    "Burn the jewish"
]
```

```python
inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=512)
```

```python
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits
```

```python
probabilities = torch.softmax(logits, dim=-1)
```

```python
# Afficher les résultats
for i, text in enumerate(texts):
    print(f"Texte: {text}")
    print(f"Probabilité (non toxique, toxique) : {probabilities[i].numpy()}")
    predicted_class = torch.argmax(probabilities[i]).item()
    if predicted_class == 1:
        print("Classifié comme toxique")
    else:
        print("Classifié comme non toxique")
    print("-" * 50)
```

```python
# Charger le modèle et le tokenizer Jigsaw Multilingual Toxic Comment Classifier
model_name = "jigsaw-multilingual-toxic-comment-classifier"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=512)
```

```python
# Tokeniser les textes
inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=512)

# Faire des prédictions avec le modèle
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

# Convertir les logits en probabilités (car il s'agit d'une classification multi-classes)
probabilities = torch.sigmoid(logits)
```

```python
# Les catégories de toxicité prises en compte par ce modèle
categories = ['toxique', 'insultes', 'menaces', 'langage haineux', 'sexisme', 'racisme']

# Afficher les résultats pour chaque texte
for i, text in enumerate(texts):
    print(f"Texte : {text}")
    print("Probabilités pour chaque catégorie :")
    for j, category in enumerate(categories):
        print(f"  {category}: {probabilities[i][j].item():.2f}")
    
    # Affichage des catégories considérées comme toxiques (seuil de 0.5)
    toxic_categories = [categories[j] for j in range(len(categories)) if probabilities[i][j].item() >= 0.5]
    
    if toxic_categories:
        print(f"Classé comme toxique dans les catégories: {', '.join(toxic_categories)}")
    else:
        print("Classé comme non toxique")
    
    print("-" * 50)
```
